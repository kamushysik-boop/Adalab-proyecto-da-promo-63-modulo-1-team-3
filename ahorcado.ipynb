```python
ahorcado = [
    """
     -----
     |   |
         |
         |
         |
         |
    --------
    """,
    """
     -----
     |   |
     O   |
         |
         |
         |
    --------
    """,
    """
     -----
     |   |
     O   |
     |   |
         |
         |
    --------
    """,
    """
     -----
     |   |
     O   |
    /|   |
         |
         |
    --------
    """,
    """
     -----
     |   |
     O   |
    /|\\  |
         |
         |
    --------
    """,
    """
     -----
     |   |
     O   |
    /|\\  |
    /    |
         |
    --------
    """,
    """
     -----
     |   |
     O   |
    /|\\  |
    / \\  |
         |
    --------
    """
]

# ===============================
# BIENVENIDA
# ===============================
print("üéÆ ===============================")
print("üéÆ   BIENVENIDO AL AHORCADO")
print("üéÆ ===============================\n")

print("üë• C√≥mo se juega:")
print("- El Jugador 1 escribe una palabra secreta.")
print("- El Jugador 2 debe adivinarla letra por letra.")
print("- Cada fallo ir√° completando el dibujo del ahorcado.")
print("- Si el dibujo se completa, el Jugador 2 pierde.")
print("- Si descubre la palabra a tiempo, gana.\n")

print("üëâ Jugador 1: escribe tu palabra oculta")
print("‚ö†Ô∏è Aseg√∫rate de que el Jugador 2 NO vea la pantalla.\n")

# ===============================
# ELECCI√ìN DE PALABRA (JUGADOR 1)
# ===============================
while True:
    palabra_oculta = input("Jugador 1, introduce la palabra secreta: ").lower()

    palabra_oculta = "".join(c for c in palabra_oculta if c.isalpha())

    if palabra_oculta:
        break
    else:
        print("üö´ La palabra solo debe contener letras (sin espacios ni n√∫meros).")



print("\n" * 50)
print("üîí Palabra guardada correctamente.")
print("üëâ Turno del Jugador 2. ¬°Empieza el juego!\n")

# ===============================
# INICIALIZACI√ìN
# ===============================
palabra_mostrada = ["_"] * len(palabra_oculta)
letras_usadas = set()
fallos = 0
salida_voluntaria = False


# ===============================
# EXCEPCI√ìN PERSONALIZADA
# ===============================
class ErrorLetra(Exception):
    pass


def adivinar_palabra(letra):
    try:
        letra = letra.strip().lower()

        if not letra.isalpha() or len(letra) != 1:
            raise ErrorLetra("Introduce una √∫nica letra v√°lida (a-z)")

        if letra in letras_usadas:
            print("‚ö†Ô∏è Ya intentaste esa letra.")
            return None

        letras_usadas.add(letra)
        acierto = False

        for i in range(len(palabra_oculta)):
            if palabra_oculta[i] == letra:
                palabra_mostrada[i] = letra
                acierto = True

        if acierto:
            print("‚úÖ La letra est√° en la palabra.")
            return True
        else:
            print("‚ùå La letra no est√° en la palabra.")
            return False

    except ErrorLetra as e:
        print("üö´", e)
        return None


# ===============================
# BUCLE PRINCIPAL DEL JUEGO
# ===============================
while fallos < len(ahorcado) - 1 and "_" in palabra_mostrada:
    print("\nPalabra:", " ".join(palabra_mostrada))

    if letras_usadas:
        print("Letras usadas:", " ".join(sorted(letras_usadas)))
    else:
        print("Letras usadas: ninguna")

    print("Fallos:", fallos)
    print(ahorcado[fallos])

    letra_usuario = input("Jugador 2, introduce una letra (o escribe 'salir'): ").strip().lower()

    if letra_usuario == "salir":
        salida_voluntaria = True
        break

    resultado = adivinar_palabra(letra_usuario)

    if resultado is None:
        print("‚ö†Ô∏è Entrada inv√°lida o repetida. Prueba otra.")
        continue

    if resultado is False:
        fallos += 1


# ===============================
# FIN DEL JUEGO
# ===============================
if salida_voluntaria:
    print("\nüëã El Jugador 2 ha salido del juego.")
    print("La palabra era:", palabra_oculta)

elif "_" not in palabra_mostrada:
    print("\nüéâ ¬°El Jugador 2 ha ganado!")
    print("La palabra era:", palabra_oculta)

else:
    print("\nüíÄ El Jugador 2 ha perdido")
    print(ahorcado[-1])
    print("Fallos:", fallos)
    print("La palabra era:", palabra_oculta)
```
