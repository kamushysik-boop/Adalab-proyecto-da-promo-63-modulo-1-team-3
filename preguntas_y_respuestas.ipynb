```python
def pregunta_aleatoria(preguntas):                                   # funcion que sirve para obtener una pregunta aleatoria del diccionario de preguntas
    return random.choice(list(preguntas.keys()))
```

```python
def jugar(preguntas):
    correctas = 0
    incorrectas = 0

    print("Â¡Comienza el juego de Preguntas y Respuestas de GeografÃ­a! \nGanas al responder 5 preguntas correctamente. Pierdes al responder 3 preguntas incorrectamente.\n")

    while correctas < 5 and incorrectas < 3:
        pregunta = pregunta_aleatoria(preguntas)

        respuesta_jugador = input(pregunta).lower()
        print(f"Has respondido: {respuesta_jugador}")

        if respuesta_jugador == preguntas[pregunta]:
            correctas += 1
            print("Correcto!")

        elif respuesta_jugador.lower() == "salir":
            break

        else:
            incorrectas += 1
            print("Respuesta incorrecta.")

        print(f"Puntaje: Correctas: {correctas} | Incorrectas: {incorrectas}\n")            # Recordar usar f string cuando queremos mostrar otras variables dentro del string

    if correctas == 5:
        return "ğŸ‰5 respuestas correctas. Â¡Ganaste!ğŸ‰"

    elif incorrectas == 3:
        return "âŒ3 respuestas incorrectas. Fin del juego.âŒ"

    else:
        return "Has salido del juego."
```
