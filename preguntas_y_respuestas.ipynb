## Juego Preguntas y Respuestas

```python
import random
```

```python
preguntas_geografia = {"¿Capital de Francia?": "paris",
    "¿En qué continente está Brasil?": "america",
    "¿Capital de Argentina?": "buenos aires",
    "¿Cuál es el río más largo del mundo?": "amazonas",
    "¿Capital de España?": "madrid",
    "¿Capital de Italia?": "roma",
    "¿Cuál es el océano más grande del mundo?": "pacifico",
    "¿Capital de Perú?": "lima",
    "¿Capital de Chile?": "santiago",
    "¿Qué país tiene forma de bota?": "italia"}
```

```python
def pregunta_aleatoria(preguntas):
    return random.choice(list(preguntas.keys()))
```

```python
pregunta_aleatoria(preguntas_geografia)
```
