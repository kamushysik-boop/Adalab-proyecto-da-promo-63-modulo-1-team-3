```python
import random

puntos_para_ganar = 3

def obtener_eleccion_jugador():
    # 1. CORRECCIÃ“N: Usamos un bucle para que pregunte hasta que sea correcto
    while True:
        eleccion = input("Elige piedra, papel o tijera: ").lower().strip()
        if eleccion in ["piedra", "papel", "tijera"]:
            return eleccion
        print("âš ï¸ OpciÃ³n no vÃ¡lida. IntÃ©ntalo de nuevo.")

def jugar_vs_pc():
    puntos_jugador = 0
    puntos_pc = 0
    opciones = ["piedra", "papel", "tijera"]

    print("\n--- ğŸ‘¤ JUGADOR VS ğŸ’» ORDENADOR ---")
    print(f"El primero en llegar a {puntos_para_ganar} puntos gana.\n")

    while puntos_jugador < puntos_para_ganar and puntos_pc < puntos_para_ganar: 
        eleccion_jugador = obtener_eleccion_jugador()
        eleccion_pc = random.choice(opciones)
        print(f"ğŸ’» El ordenador eligiÃ³: {eleccion_pc}")

        if eleccion_jugador == eleccion_pc:
            print("ğŸ¤ Â¡Es un empate!")
        elif (eleccion_jugador == "piedra" and eleccion_pc == "tijera") or \
             (eleccion_jugador == "tijera" and eleccion_pc == "papel") or \
             (eleccion_jugador == "papel" and eleccion_pc == "piedra"):
            print("âœ¨ Â¡Ganaste esta ronda!")
            puntos_jugador += 1
        else:
            print("ğŸ¤– El ordenador gana esta ronda.")
            puntos_pc += 1

        print(f"Marcador: ğŸ‘¤ {puntos_jugador} - ğŸ’» {puntos_pc}\n")

    if puntos_jugador == puntos_para_ganar:
        print("*******************************")
        print("Â¡FELICIDADES! Has ganado el juego. ğŸ†")
        print("*******************************\n")
    else:
        print("*******************************")
        print("El ordenador gana el juego. â˜¹ï¸")
        print("*******************************\n")

def juego_multijugador(): 
    puntos_jugador1 = 0
    puntos_jugador2 = 0
    lista_de_opciones = ("piedra","papel","tijera")

    print("\n--- ğŸ”µ JUGADOR 1 VS ğŸ”´ JUGADOR 2 ---")

    while puntos_jugador1 < 3 and puntos_jugador2 < 3:
        j1 = input("ğŸ”µ Jugador 1: ").lower().strip()
        j2 = input("ğŸ”´ Jugador 2: ").lower().strip()
        
        if j1 not in lista_de_opciones or j2 not in lista_de_opciones:
            print("âš ï¸ ELECCIÃ“N NO VÃLIDA.")
            continue

        print(f"Elecciones: ğŸ”µ {j1} | ğŸ”´ {j2}")

        if j1 == j2:
            print("ğŸ¤ EMPATE")
        elif (j1 == "piedra" and j2 == "tijera") or \
             (j1 == "papel" and j2 == "piedra") or \
             (j1 == "tijera" and j2 == "papel"):
            puntos_jugador1 += 1
            print("â­ Gana Jugador 1")
        else:
            puntos_jugador2 += 1
            print("â­ Gana Jugador 2")
        
        print(f"Marcador: ğŸ”µ {puntos_jugador1} | ğŸ”´ {puntos_jugador2}\n")

    # 2. CORRECCIÃ“N: Usar print en lugar de return para que el cliente lo vea
    if puntos_jugador1 == 3:
        print("*******************************")
        print("ğŸ”µ Â¡JUGADOR 1 GANA EL JUEGO! ğŸ†")
        print("*******************************\n")
    else:
        print("*******************************")
        print("ğŸ”´ Â¡JUGADOR 2 GANA EL JUEGO! ğŸ†")
        print("*******************************\n")
 
def menu():
    while True:
        print("==========================")
        print("   MENÃš PRINCIPAL ğŸ®")
        print("==========================")
        print("1. Jugador vs Ordenador")
        print("2. Multijugador")
        print("3. Salir")

        opcion = input("Elige 1, 2 o 3: ")

        if opcion == "1":
            jugar_vs_pc()
        elif opcion == "2":
            juego_multijugador()
        elif opcion == "3":
            print("Â¡Gracias por jugar! ğŸ‘‹")
            break
        else:
            print("âŒ OpciÃ³n no vÃ¡lida.\n")

if __name__ == "__main__":
    menu()
```
