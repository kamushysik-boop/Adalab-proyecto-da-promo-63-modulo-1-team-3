```python
import random
puntos_para_ganar = 3
def obtener_eleccion_jugador():
    eleccion = input("Elige Piedra, Papel o Tijera: ").lower()
    if eleccion in ["piedra", "papel", "tijera"]:
        return eleccion
    else:
        eleccion = input("OpciÃ³n no vÃ¡lida. Por favor, escribe piedra, papel o tijera: ").lower()

def jugar_vs_pc():
    puntos_jugador = 0
    puntos_pc = 0
    opciones = ["piedra", "papel", "tijera"]

    print(emoji.emojize("Â¡Bienvenido al juego de :puÃ±o: :mano_abierta: o :tijeras:!", language='es'))
    print(f"El primero en llegar a {puntos_para_ganar} puntos gana.\n")

    while puntos_jugador < puntos_para_ganar and puntos_pc < puntos_para_ganar:
        #print(f"\nMarcador â†’ Jugador: {puntos_jugador} | PC: {puntos_pc}")

        eleccion_jugador = obtener_eleccion_jugador()
        eleccion_pc = random.choice(opciones)
        print(f"El ordenador eligiÃ³: {eleccion_pc}")

        if eleccion_jugador == eleccion_pc:
            print("Â¡Es un empate!")
        elif (
            (eleccion_jugador == "piedra" and eleccion_pc == "tijera") or 
            (eleccion_jugador == "tijera" and eleccion_pc == "papel") or
            (eleccion_jugador == "papel" and eleccion_pc == "piedra")):
            print("Â¡Ganaste esta ronda!")
            puntos_jugador += 1
        else:
            print("El ordenador gana esta ronda.")
            puntos_pc += 1

        print(f"Marcador: TÃº {puntos_jugador} - PC {puntos_pc}\n")

    if puntos_jugador == puntos_para_ganar:
        print("\n*******************************")
        print("Â¡FELICIDADES! Has ganado el juego.")
        print("*******************************\n")
    else:
        print("\n*******************************")
        print("Lo siento, el ordenador ha ganado el juego. Â¡Sigue practicando!")
        print("*******************************\n")

def juego_multijugador(): 
    puntos_jugador1  = 0
    puntos_jugador2  = 0
    puntos_a_ganar  = 3
    lista_de_opciones = ("piedra","papel","tijera")

    print("Â¡Bienvenido al juego de Piedra, Papel o Tijera!")
    print(f"El primero en llegar a {puntos_a_ganar} puntos gana.\n")

    while puntos_jugador1 < puntos_a_ganar and puntos_jugador2 < puntos_a_ganar:
        jugador1 = input("ELIGE PIEDRA,PAPEL,TIJERA").lower()
        jugador2 = input("ELIGE PIEDRA,PAPEL,TIJERA").lower()

        print(f"Jugador1 ha elegido:--> {jugador1}    Jugador2 ha elegido:--> {jugador2}")  

        if jugador1 not in lista_de_opciones or jugador2 not in lista_de_opciones:
            print("ELECCIÃ“N NO VALIDA, SOLO SE PUEDE ESCOGER PIEDRA,PAPEL O TIJERA")
        elif jugador1 == jugador2:
            print("EMPATE")
        elif jugador1 == "piedra" and jugador2 == "tijera":
            puntos_jugador1 =  puntos_jugador1 + 1
            print(f"GANA JUGADOR1 ---> Marcador: Jugador1 {puntos_jugador1} Jugador2 {puntos_jugador2}" )
        elif jugador1 == "papel" and jugador2 == "piedra":
            puntos_jugador1 = puntos_jugador1 + 1
            print(f"GANA JUGADOR1 ---> Marcador: Jugador1 {puntos_jugador1} Jugador2 {puntos_jugador2}")
        elif jugador1 == "tijera" and jugador2 == "papel":
            puntos_jugador1 = puntos_jugador1 + 1
            print(f"GANA JUGADOR1 ---> Marcador: Jugador1 {puntos_jugador1} Jugador2 {puntos_jugador2}")     
        else:
            puntos_jugador2 = puntos_jugador2 + 1
            print(f"GANA JUGADOR2 ---> Marcador: Jugador1 {puntos_jugador1} Jugador2 {puntos_jugador2}") 
       
    if puntos_jugador1 == 3:
        print("\n*******************************")
        print("JUGADOR1 ERES EL GANADOR")
        print("*******************************\n")
    else:
        print("\n*******************************")
        print("JUGADOR2 ERES EL GANADOR")
        print("*******************************\n")
 
def menu():
    while True:
        print("==========================")
        print("Seleccione modo de juego:")
        print("==========================")
        print("1. Jugador vs Ordenador")
        print("2. Multijugador")
        print("3. Salir")

        opcion = input("Elige 1, 2 o 3: ")

        if opcion == "1":
            jugar_vs_pc()
            print(f"Has seleccionado jugar contra el ordenador")
        elif opcion == "2":
            juego_multijugador()
            print(f"Has seleccionado jugar contra otro jugador")
        elif opcion == "3":
            print("Â¡Gracias por jugar! ðŸ‘‹")
            break
        else:
            print("OpciÃ³n no vÃ¡lida.\n")

if __name__ == "__main__":
    menu()
 
```
