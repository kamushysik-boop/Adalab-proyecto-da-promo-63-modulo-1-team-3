# Piedra Papel & Tijera

- Reglas del Juego "Piedra, Papel y Tijera":

    - Dos jugadores eligen una de las tres opciones: "Piedra", "Papel" o "Tijera".

    - Las reglas para determinar al ganador son:
        
        - Piedra vence a Tijera.

        - Tijera vence a Papel.

        - Papel vence a Piedra.

        - El jugador que gana una ronda obtiene un punto.
        
        - El juego continúa hasta que un jugador haya acumulado 3 puntos, convirtiéndose en el ganador.


- Pistas para las alumnas:

    1. Comienza por definir el número de puntos objetivo que se deben alcanzar para ganar el juego.

    2. Implementa una función que obtenga la elección del jugador y valide que sea una de las opciones válidas ("Piedra", "Papel" o "Tijera").

    3. Utiliza un bucle `while` para llevar un registro de los puntos de cada jugador y continuar el juego hasta que se alcance el número de puntos objetivo.

    4. Genera una elección aleatoria para el segundo jugador (la computadora) utilizando el módulo `random`.

    5. Determina al ganador de cada ronda y actualiza los puntos correspondientes.

    6. Verifica si algún jugador ha alcanzado el número de puntos objetivo y finalicen el juego cuando eso suceda.


```python


def juego_multijugador(puntos_jugador1 , puntos_jugador2): 
 puntos_jugador1  = 0
 puntos_jugador2  = 0
 puntos_a_ganar  = 3
 #uso una tupla que no puede ser modificada
 lista_de_opciones = ("piedra","papel","tijera")

 while puntos_jugador1 < puntos_a_ganar and puntos_jugador2 < puntos_a_ganar:
#eleciones de los jugadores
   jugador1 = input("ELIGE PIEDRA,PAPEL,TIJERA").lower()
   jugador2 = input("ELIGE PIEDRA,PAPEL,TIJERA").lower()
   #imprimimos para recordar las elecciones de cada jugador 
   print(f"Jugador1 ha elegido:--> {jugador1}    Jugador2 ha elegido:--> {jugador2}")
   #para que solo me de como validas las tres opciones,si no te valida cuanquier palabra  
   if jugador1 not in lista_de_opciones or jugador2 not in lista_de_opciones:
        print("ELECCIÓN NO VALIDA, SOLO SE PUEDE ESCOGER PIEDRA,PAPEL O TIJERA")
   elif jugador1 == jugador2:
        print("EMPATE")
   elif jugador1 == "piedra" and jugador2 == "tijera":
        #suma de cada vuelta 1
        puntos_jugador1 =  puntos_jugador1 + 1
        print(f"GANA JUGADOR1 ---> Marcador: Jugador1 {puntos_jugador1} Jugador2 {puntos_jugador2}" )
   elif jugador1 == "papel" and jugador2 == "piedra":
        #suma de cada vuelta 1
        puntos_jugador1 = puntos_jugador1 + 1
        print(f"GANA JUGADOR1 ---> Marcador: Jugador1 {puntos_jugador1} Jugador2 {puntos_jugador2}")
   elif jugador1 == "tijera" and jugador2 == "papel":
        #suma de cada vuelta 1
        puntos_jugador1 = puntos_jugador1 + 1
        print(f"GANA JUGADOR1 ---> Marcador: Jugador1 {puntos_jugador1} Jugador2 {puntos_jugador2}")     
   #si ninguna de las anteriores se cumple entonces ganaria jugador2
   else:
        puntos_jugador2 = puntos_jugador2 + 1
        print(f"GANA JUGADOR2 ---> Marcador: Jugador1 {puntos_jugador1} Jugador2 {puntos_jugador2}") 
 #salimos del while para ver quien consigio 3 puntos       
 if puntos_jugador1 == 3:
     return "JUGADOR1 ERES EL GANADOR"
 else:
     return "JUGADOR2 ERES EL GANADOR"
```
