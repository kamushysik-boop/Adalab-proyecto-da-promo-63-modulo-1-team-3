```python
import random 
try:
   def obtener_eleccion_jugador():
     while True:
        eleccion = input("Elige Piedra, Papel o Tijera: ").lower()
        if eleccion in ["piedra", "papel", "tijera"]:
            return eleccion
except:
   print("Opción no válida. Por favor, escribe piedra, papel o tijera: ")



def jugar_contra_ordenador():
    puntos_jugador = 0
    puntos_ordenador = 0
    puntos_para_ganar = 3
    ordenador = random.choice(["piedra", "papel", "tijera"])

   

    # 2. Función para obtener la elección del jugador

    while puntos_jugador < puntos_para_ganar and puntos_ordenador < puntos_para_ganar: 
        print(f"\nMarcador → Jugador: {puntos_jugador} | Computadora: {puntos_ordenador}")

        jugador = obtener_eleccion_jugador()
        ordenador = random.choice(["piedra", "papel", "tijera"])

        if jugador == ordenador:
            print("¡Es un empate!")
        elif (jugador == "piedra" and ordenador == "tijera") or \
             (jugador == "tijera" and ordenador == "papel") or \
             (jugador == "papel" and ordenador == "piedra"):
            print("¡Ganaste esta ronda!")
            puntos_jugador += 1
        else:
            print("La computadora gana esta ronda.")
            puntos_ordenador += 1

        print(f"Marcador: Tú {puntos_jugador} - PC {puntos_ordenador}\n")

    if puntos_jugador == puntos_para_ganar:
            
            print("¡FELICIDADES! Has ganado el juego.")
    else:
        print("Lo siento, el ordenador ha ganado el juego. ¡Sigue practicando!")





    
```

```python
def juego_multijugador(puntos_jugador1 , puntos_jugador2): 
 puntos_jugador1  = 0
 puntos_jugador2  = 0
 puntos_para_ganar = 3
 #uso una tupla que no puede ser modificada
 lista_de_opciones = ("piedra","papel","tijera")

 while puntos_jugador1 < puntos_para_ganar and puntos_jugador2 < puntos_para_ganar:
#eleciones de los jugadores
   jugador1 = input("ELIGE PIEDRA,PAPEL,TIJERA").lower()
   jugador2 = input("ELIGE PIEDRA,PAPEL,TIJERA").lower()
   #imprimimos para recordar las elecciones de cada jugador 
   print(f"Jugador1 ha elegido:--> {jugador1}    Jugador2 ha elegido:--> {jugador2}")
   #para que solo me de como validas las tres opciones,si no te valida cuanquier palabra  
   if jugador1 not in lista_de_opciones or jugador2 not in lista_de_opciones:
        print("Opción no válida. Por favor, escribe piedra, papel o tijera")
   elif jugador1 == jugador2:
        print("¡Es un empate!")
   elif jugador1 == "piedra" and jugador2 == "tijera":
        #suma de cada vuelta 1
        puntos_jugador1 =  puntos_jugador1 + 1
        print(f" GANA JUGADOR1 ---> Marcador: Jugador1 {puntos_jugador1} Jugador2 {puntos_jugador2}" )
   elif jugador1 == "papel" and jugador2 == "piedra":
        #suma de cada vuelta 1
        puntos_jugador1 = puntos_jugador1 + 1
        print(f"GANA JUGADOR1 ---> Marcador: Jugador1 {puntos_jugador1} Jugador2 {puntos_jugador2}")
   elif jugador1 == "tijera" and jugador2 == "papel":
        #suma de cada vuelta 1
        puntos_jugador1 = puntos_jugador1 + 1
        print(f"GANA JUGADOR1 ---> Marcador: Jugador1 {puntos_jugador1} Jugador2 {puntos_jugador2}")     
   #si ninguna de las anteriores se cumple entonces ganaria jugador2
   else:
        puntos_jugador2 = puntos_jugador2 + 1
        print(f"GANA JUGADOR2 ---> Marcador: Jugador1 {puntos_jugador1} Jugador2 {puntos_jugador2}") 
 #salimos del while para ver quien consigio 3 puntos       
 if puntos_jugador1 == 3:
     return "JUGADOR1 , ¡FELICIDADES! Has ganado el juego."
 else:
     return "JUGADOR2 , ¡FELICIDADES! Has ganado el juego."
```
